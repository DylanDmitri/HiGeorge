````python

class Building:
    def load(self, row):
        self.zipcode = row[3]
        self.address = row[4]
        self.posted_at = row[12][:10]
        self.company = row[14]
        
        self.prices = [int(row[9])]  # price
    
    def update(self, row, writer):
        if len(row) < 15:
            return
        
        try:
            # filters out *most* poorly formed lines from bad quotes
            price = int(row[9])
        except:
            return
        
        if row[3] == self.zipcode and row[4] == self.address:
            self.prices.append(int(row[9]))  # part of same building
        else:
            writer.writerow([self.zipcode, self.address.lower(), self.posted_at, self.company.lower(),
                            len(self.prices), sum(self.prices)])
            self.load(row)
            
processed = 1
with gzip.open('renthub_data.csv.gz', 'rt') as f:
    reader = csv.reader(f)
    writer = csv.writer(open('renthub_data.csv', 'w', newline=''))

    building = Building()
    next(reader)
    building.load(next(reader))
    
    while True:
        try:
            row = next(reader)
        except (UnicodeDecodeError, IndexError) as e:
            continue
            
        building.update(row, writer)
        
        processed += 1
        if processed % 1_000_000 == 0:
            print(processed)
````

Next step is merging buildings having same (zip + address + posted MONTH + company)

````python
merged = {}

with open('renthub_data_cleaned.csv', 'r') as f:
    reader = csv.reader(f)
    for line in reader:
        zipcode, address, posted_at, company, rooms, tot_price = line
        try:
            rooms = int(rooms)
            tot_price = int(tot_price)
        except ValueError:
            rooms = 0
            tot_price = 0
        key = (zipcode, address, posted_at[:7], company)

        if key in merged:
            merged[key][0] += rooms
            merged[key][1] += tot_price
        else:
            merged[key] = [rooms, tot_price]
````

## deduplicate

If multiple companies have same (zip + address + month posted) keep the one with more rooms.

````python
buildings = {}
with open('renthub_data_merged.csv') as f:
    
    for line in csv.reader(f):
        zipcode, address, posted_at, company, rooms, tot_price = line
        rooms = int(rooms)
        tot_price = int(tot_price)
        
        key = zipcode + address + posted_at
        if key in buildings:
            prev_line = buildings[key]
            prev_rooms = int(prev_line[4])
            if rooms > prev_rooms:
                buildings[key] = line
        else:
            buildings[key] = line
        
````

## determine insufficient data

````python
from collections import defaultdict

target = '98052'
threshold = -1.9  # sigma below this get flagged

rooms_by_month = defaultdict(lambda: 0)

for line in buildings.values():
    zipcode = line[0]
    if zipcode != target:
        continue
        
    month = line[2]
    rooms = int(line[4])
    
    rooms_by_month[month] += rooms
    
rooms_by_month
````

````python
volume = np.array(list(rooms_by_month.values()))
mean, dev = np.mean(volume), np.std(volume) 

deviations_off = (volume - mean) / dev
deviations_off
````

````python
problem_months = deviations_off < threshold
problem_months = dict(zip(rooms_by_month.keys(), problem_months))
for key, val in problem_months.items():
    if val:
        print(key, 'has unusually low volume')
````

## price analysis for 98052

````python
target = '98052'

tot_rooms = 0
tot_price = 0

class Averager:
    def __init__(self):
        self.rooms = 0
        self.price = 0
    
    @property
    def average(self):
        return self.price / self.rooms

prices_over_time = defaultdict(lambda: Averager())

for line in buildings.values():
    zipcode = line[0]
    if zipcode != target:
        continue

    month = line[2]
    prices_over_time[month].rooms += int(line[4])
    prices_over_time[month].price += int(line[5])

for key, val in sorted(prices_over_time.items()):
    print(key, val.average)
````

````python
import matplotlib.pyplot as plt

tups = sorted(prices_over_time.items())
date = [t[0] for t in tups]
price = [t[1].average for t in tups]

plt.plot(date, price)
plt.xticks(rotation=45)
plt.show()
````
